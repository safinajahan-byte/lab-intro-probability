**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
import scipy.stats as stats

# Parameters
n = 460  # Tickets sold
p_miss = 0.03
p_show = 1 - p_miss  # 0.97 probability of showing up
seats = 450

# Calculation
# We want the probability that 450 or fewer people show up
probability = stats.binom.cdf(seats, n, p_show)

print(f"The probability that everyone gets a seat is: {probability:.4f}")
print(f"Percentage: {probability * 100:.2f}%")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
import scipy.stats as stats

p = 0.3

# Method A: Using the formula for "More than k"
# P(X > k) = (1 - p)^k
# We want at least 3, which means we must fail the first 2.
prob_at_least_3_manual = (1 - p)**2

# Method B: Using Scipy's Survival Function (SF)
# sf(k) = P(X > k). Since we want X >= 3, that's the same as X > 2.
prob_at_least_3_scipy = stats.geom.sf(2, p)

print(f"Probability of needing at least 3 attempts: {prob_at_least_3_scipy:.4f}")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
import scipy.stats as stats

# Average visits (mu or lambda)
avg_visits = 500
capacity = 550

# Method 1: 1 - CDF
prob_overwhelmed = 1 - stats.poisson.cdf(capacity, avg_visits)

# Method 2: Survival Function (SF) - This is mathematically 1 - CDF
prob_overwhelmed_sf = stats.poisson.sf(capacity, avg_visits)

print(f"Probability of server being overwhelmed: {prob_overwhelmed_sf:.4f}")
print(f"Percentage: {prob_overwhelmed_sf * 100:.2f}%")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
import scipy.stats as stats

# Probability for a single hour
p_hour = stats.poisson.sf(550, 500)

# Probability for at least once in 24 hours
p_day = 1 - (1 - p_hour)**24

print(f"Probability of being overwhelmed at least once in 24h: {p_day:.4f}")
print(f"Percentage: {p_day * 100:.2f}%")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
import scipy.stats as stats

# Average time between arrivals (1/lambda)
avg_time = 10 

# Target arrival time
target_time = 5

# Using the scale parameter (scale = 1/lambda)
prob_within_5 = stats.expon.cdf(target_time, scale=avg_time)

print(f"Probability of arrival within 5 minutes: {prob_within_5:.4f}")
print(f"Percentage: {prob_within_5 * 100:.2f}%")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
import scipy.stats as stats

avg_time = 10
threshold = 15

# Method 1: 1 - CDF
prob_break_cdf = 1 - stats.expon.cdf(threshold, scale=avg_time)

# Method 2: Survival Function (More precise)
prob_break_sf = stats.expon.sf(threshold, scale=avg_time)

print(f"Probability of no customer for 15 minutes: {prob_break_sf:.4f}")
print(f"Percentage chance of a break: {prob_break_sf * 100:.2f}%")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
import scipy.stats as stats

mean = 150
std_dev = 10

# Calculate the area to the left of 160 and subtract the area to the left of 140
prob_between = stats.norm.cdf(160, mean, std_dev) - stats.norm.cdf(140, mean, std_dev)

print(f"Probability weight is between 140g and 160g: {prob_between:.4f}")
print(f"Percentage: {prob_between * 100:.2f}%")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
import scipy.stats as stats

# Parameters
mean_lifetime = 50 # this is our 'scale'
time_limit = 30

# Calculate the CDF for 30 hours
prob_failure = stats.expon.cdf(time_limit, scale=mean_lifetime)

print(f"Probability of failure within 30 hours: {prob_failure:.4f}")
print(f"Percentage: {prob_failure * 100:.2f}%")
```
